## Hyperparamters

```python
BATCH_SIZE = 16

EPOCHS = 20
IMAGE_X, IMAGE_Y = 64, 64
NUM_DAYS_PER_SAMPLE = 7
```

```python
from att_bott_utils import load_sequential_data, create_flow, create_att_bottleneck_model, tf
```

# Unscaled Maps

## Load Model from Module

```python
# Global Arguments
MODEL_ARCHITECTURE = (16, 16, 16)
SEQUENCE_LENGTH = NUM_DAYS_PER_SAMPLE
NUM_HEADS = 8
NUM_PAD_TOKENS = 64
JOIN_LAYER = 2
REFRESH_PAD_TOKENS = False
D_MODEL = 32
OUTPUT_SIZE = 49

# Image Arguments
IMAGE_DIMS = {
    "0": IMAGE_Y,
    "1": IMAGE_X,
    "2": 1
}
KERNEL_SIZE = (4, 4)
MAXPOOL_KERNEL = 2

# Graph Arguments
INPUT_SHAPE_NODES = (NUM_DAYS_PER_SAMPLE, OUTPUT_SIZE, 5)
INPUT_SHAPE_EDGES = (NUM_DAYS_PER_SAMPLE, OUTPUT_SIZE, OUTPUT_SIZE)
SEQUENCE_LENGTH_GRAPH = 49
RESIDUAL = True
USE_BIAS = True
```

```python
att_lstm = create_att_bottleneck_model(
    layer_units = MODEL_ARCHITECTURE,
    sequence_length = SEQUENCE_LENGTH,
    num_heads = NUM_HEADS,
    num_pad_tokens = NUM_PAD_TOKENS,
    join_layer = JOIN_LAYER,
    refresh_pad_tokens = REFRESH_PAD_TOKENS,
    d_model = D_MODEL,
    output_size = OUTPUT_SIZE,
    image_dims = IMAGE_DIMS,
    kernel_size = KERNEL_SIZE,
    maxpool_kernel = MAXPOOL_KERNEL,
    input_shape_nodes = INPUT_SHAPE_NODES,
    input_shape_edges = INPUT_SHAPE_EDGES,
    sequence_length_graph = SEQUENCE_LENGTH_GRAPH,
    residual = RESIDUAL,
    use_bias = USE_BIAS,
    name = "AttentionBottleneckLSTM"
)
```

```python
att_lstm.summary()
```

## Load Sequential Data

```python
(X_indices, y_infection, y_death, X_maps), (formatted_X, formatted_adj_mat, formatted_y_infection, formatted_y_death) = load_sequential_data(
    maps_path = unscaled_maps_path,
    metadata_path = unscaled_metadata_path,
    covid_data_path = dataset_path,
    flight_data_path = flight_path,
    image_x = IMAGE_X,
    image_y = IMAGE_Y,
    num_days_per_sample = NUM_DAYS_PER_SAMPLE
)
```

## Scale and Split Train Test

```python
from sklearn.model_selection import train_test_split
```

```python
# MinMax Scale by column
X_max = formatted_X.reshape(-1, formatted_X.shape[-1]).max(axis = 0)
X_min = formatted_X.reshape(-1, formatted_X.shape[-1]).min(axis = 0)
X_nodes_scaled = (formatted_X - X_min) / (X_max - X_min)

mat_max = formatted_adj_mat.reshape(-1,).max(axis = 0)
mat_min = formatted_adj_mat.reshape(-1,).min(axis = 0)
edges_scaled = (formatted_adj_mat - mat_min) / (mat_max - mat_min)

X_nodes_scaled.shape, edges_scaled.shape
```

```python
(X_nodes_train, X_nodes_test,
X_edges_train, X_edges_test,
X_map_ind_train, X_map_ind_test,
y_train, y_test) = train_test_split(
    X_nodes_scaled,
    edges_scaled,
    X_indices,
    y_infection,
    train_size = .8,
    random_state = 0
)
```

## Train Model

```python
opt = tf.keras.optimizers.Adam(learning_rate = 0.001)

@tf.function
def loss_func(y_true, y_pred):
  squared_diff = tf.square(y_true - y_pred)
  mse = tf.reduce_mean(squared_diff, axis=-1)
  squared_diff_mean = tf.square(tf.reduce_mean(y_true, axis = -1) - tf.reduce_mean(y_pred, axis = -1))
  return mse + squared_diff_mean

att_lstm.compile(
    loss = loss_func,
    optimizer = opt,
    metrics = ["mse", "mae"]
)
```

```python
train_generator = create_flow(
    BATCH_SIZE,
    X_map_ind_train,
    X_maps,
    X_nodes_train,
    X_edges_train,
    y_train
)
```

```python
STEPS_PER_EPOCH = X_map_ind_train.shape[0] // BATCH_SIZE
```

```python
history = att_lstm.fit(
    x = train_generator,
    batch_size = BATCH_SIZE,
    steps_per_epoch = STEPS_PER_EPOCH,
    epochs = EPOCHS
    )
```

## Evaluate Model

```python
test_generator = create_flow(
    BATCH_SIZE,
    X_map_ind_test,
    X_maps,
    X_nodes_test,
    X_edges_test,
    y_test)
```

```python
STEPS = X_map_ind_test.shape[0] // BATCH_SIZE
```

```python
att_lstm.evaluate(x = test_generator, steps = STEPS, batch_size = BATCH_SIZE)
```
